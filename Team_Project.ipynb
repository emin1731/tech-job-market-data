```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
jobPostings = pd.read_csv('data/jobs.csv')
jobPostings2 = pd.read_csv('data/postings.csv')
```

```python


jobPostings = pd.read_csv('data/jobs.csv')

# jobPostings.head
job_title_counts = jobPostings["title"].value_counts()
job_title_counts = job_title_counts.sort_values(ascending=False)
top_n_titles = job_title_counts.head(10)

labels = top_n_titles.index.to_numpy()
sizes = top_n_titles.to_numpy()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Top Job Titles (In Demand)")
plt.axis("equal")  # Equal aspect ratio ensures a circular pie chart
plt.show()

```

```python
level_count = jobPostings["Seniority level"].value_counts()
level_count = level_count.sort_values(ascending=False)
top_levels = level_count.head(10)

# # Create a bar chart

# labels = top_n_titles.index.to_numpy()
# sizes = top_n_titles.to_numpy()

# # Create the pie chart
# plt.figure(figsize=(8, 8))
# plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140)


plt.figure(figsize=(10, 5))  # Adjust figure size for better readability
plt.pie(top_levels.to_numpy(), labels=top_levels.index.to_numpy(), autopct="%1.1f%%", startangle=140)  # Use index for x-axis labels
plt.xlabel("Seniority Level")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.title("Top 15 Seniority Levels (In Demand)")
plt.tight_layout()  # Adjust spacing between elements
plt.show()
```

```python
company_count = jobPostings["company"].value_counts()
company_count = company_count.sort_values(ascending=False)
top_companies = company_count.head(15)

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.bar(top_companies.index, top_companies.to_numpy())  # Use index for x-axis labels
plt.xlabel("Seniority Level")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.title("Top 15 Seniority Levels (In Demand)")
plt.tight_layout()  # Adjust spacing between elements
plt.show()
```

```python
company_count = jobPostings2["company"].value_counts()
company_count = company_count.sort_values(ascending=False)
top_companies = company_count.head(15)

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.bar(top_companies.index, top_companies.to_numpy())  # Use index for x-axis labels
plt.xlabel("Seniority Level")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.title("Top companies")
plt.tight_layout()  # Adjust spacing between elements
plt.show()
```

```python


high_salary = jobPostings["sal_high"]

low_salary = jobPostings["sal_low"]

# Create a line chart
plt.figure(figsize=(10, 5))  # Adjust figure size for better readability
plt.hist(high_salary, density=100, alpha = 0.7, bins=20, label='High salary',color='green')  # Use index for x-axis labels
plt.hist(low_salary, density=1, alpha = 0.7, bins=20, label='Low salary',color='orange')  # Use index for x-axis labels

plt.legend()
plt.xlabel("Count")
plt.ylabel("Salary")
# plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.title("Low and high salary comparison hisotgram")
plt.tight_layout()  # Adjust spacing between elements
plt.show()
```

```python

# Extract salary data
salaries = jobPostings[['sal_high', 'sal_low']]

# Create a color list for the bars (optional)
colors = ['skyblue', 'lightcoral']  # You can customize colors

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(salaries.columns, salaries.mean(), color=colors)  # Use mean for this example

# Customize the plot
plt.xlabel("Salary Range")
plt.ylabel("Average Salary ($)")  # You can use other statistics like median
plt.title("Average Salary Distribution")
plt.xticks(rotation=0)  # Optional: Prevent x-axis label rotation
plt.ylim(bottom=0)  # Set minimum y-axis value to 0 (optional)

# Add labels to bars (optional)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 50, f"{yval:.0f}", ha='center', color='black')  # Adjust offset and formatting

plt.tight_layout()
plt.show()

```
